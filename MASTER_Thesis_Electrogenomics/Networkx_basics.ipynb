```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.Graph() # normal undirected graph
```

```python
G.add_edge(1,2)
G.add_edge(2, 3, weight=0.9)
```

```python
# everything that is hashable can be a node
G.add_edge("A", "B")
G.add_edge("B", "B")
G.add_node("C")
```

```python
nx.draw_spring(G, with_labels=True)
```

```python
edge_list = [(1,2), (2,3), (3,4), (4,6), (6,7), [6,1], [3,6]]
```

```python
G = nx.from_edgelist(edge_list)
```

```python
# or to add edges from edglist:

G = nx.Graph()
G.add_edges_from(edge_list)
```

```python
# adjacency matrix

print(nx.adjacency_matrix(G))
```

```python
# adjacency matrix from numpy

ar = np.array([[0, 1, 0],
        [1,1,1],
         [0,0,0]])
```

```python
G = nx.from_numpy_array(ar)
```

```python
G
```

```python
nx.draw(G, with_labels = True)
```

```python
print(dict(G.degree))
```

```python
print(nx.shortest_path(G, 2, 6))
```
