```python
import networkx as nx

import matplotlib.pyplot as plt
```

```python
# Generate a small-world graph with 30 nodes
small_world_graph = nx.watts_strogatz_graph(30, 5, 0.3)

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(small_world_graph, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')
plt.title("Small-World Network with 30 Nodes")
plt.show()
```

```python
# Find the shortest path from node 1 to node 22
shortest_path = nx.shortest_path(small_world_graph, source=1, target=22)

# Draw the graph with the shortest path highlighted in green
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(small_world_graph, seed=33)  # Positions for all nodes

# Draw all nodes and edges
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')

# Highlight the shortest path in green
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=shortest_path, node_color='lightgreen', node_size=500)
nx.draw_networkx_edges(small_world_graph, pos, edgelist=path_edges, edge_color='green', width=2)

plt.show()
```

```python
# Redraw the graph without a title, but keeping the same layout
plt.figure(figsize=(10, 8))

# Draw the graph using the same layout and without a title
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')

plt.show()
```

```python
# Find the shortest path from node 1 to node 21 for the updated task
shortest_path_1_to_21 = nx.shortest_path(small_world_graph, source=1, target=21)

# Redraw the graph highlighting the new shortest path from node 1 to node 21
plt.figure(figsize=(10, 8))

# Draw all nodes and edges as before
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')

# Highlight the new shortest path in green
path_edges_1_to_21 = list(zip(shortest_path_1_to_21, shortest_path_1_to_21[1:]))
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=shortest_path_1_to_21, node_color='lightgreen', node_size=500)
nx.draw_networkx_edges(small_world_graph, pos, edgelist=path_edges_1_to_21, edge_color='green', width=2)

plt.show()
```

```python
# Correctly find the nodes involved in the diameter path
diameter = nx.diameter(small_world_graph)
all_shortest_paths = dict(nx.all_pairs_shortest_path(small_world_graph))
diameter_nodes = []
for source, targets in all_shortest_paths.items():
    for target, path in targets.items():
        if len(path) - 1 == diameter:
            diameter_nodes = path
            break
    if diameter_nodes:
        break

# Find the edges that make up this path
diameter_edges = list(zip(diameter_nodes, diameter_nodes[1:]))

# Redraw the graph with the diameter path highlighted
plt.figure(figsize=(10, 8))
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=diameter_nodes, node_color='orange', node_size=500)
nx.draw_networkx_edges(small_world_graph, pos, edgelist=diameter_edges, edge_color='orange', width=2)

plt.show()
```

```python
# Create a complete graph based on Example Graph 1, keeping the same number of nodes
complete_graph = nx.complete_graph(small_world_graph.number_of_nodes())

# Redraw the graph as a complete graph, using the same layout
plt.figure(figsize=(10, 8))
nx.draw(complete_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')

plt.show()
```

```python
# Re-importing necessary libraries and redefining the graph, since the code execution state was reset again
import matplotlib.pyplot as plt
import networkx as nx






# Calculate the clustering coefficient for node 17
clustering_coefficient_node_17 = nx.clustering(small_world_graph, 17)

# Get neighbours of node 17
neighbours_node_17 = list(small_world_graph.neighbors(17))

# Mark node 17 in purple and its neighbours in green
plt.figure(figsize=(10, 8))
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=[17], node_color='purple', node_size=500)
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=neighbours_node_17, node_color='green', node_size=500)

plt.show(), clustering_coefficient_node_17
```

```python
# Mark node 17 in purple, its neighbours in green, and the links between neighbours in light green, using the consistent layout

# Identifying links between the neighbors of node 17
links_between_neighbours = [(i, j) for i in neighbours_node_17 for j in neighbours_node_17 if i < j and small_world_graph.has_edge(i, j)]

# Redraw the graph with the specified nodes and links highlighted, using the consistent layout
plt.figure(figsize=(10, 8))
nx.draw(small_world_graph, pos, with_labels=True, node_color='skyblue', node_size=500, font_size=10, edge_color='gray')
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=[17], node_color='purple', node_size=500)
nx.draw_networkx_nodes(small_world_graph, pos, nodelist=neighbours_node_17, node_color='green', node_size=500)
nx.draw_networkx_edges(small_world_graph, pos, edgelist=links_between_neighbours, edge_color='lightgreen', width=2)

plt.show(), clustering_coefficient_node_17
```
